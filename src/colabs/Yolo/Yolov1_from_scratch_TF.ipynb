3. Approach(Idea)


*   Takes an image and divide it into equal size cells. (7x7 in paper) 
*   Each cell will output a bounding box

(0,0)----- x

|

y

* Output label will be **relative to cell**
* [x(0-1),y(0-1),w(<=>1),h(<=>1)]
* So, according to cell x,y represents the midpoint of the object in the cell. The w and h represents the height and width of the object relative to the cell.
* Label_cell = [c1, c2, c3, ..... c20, Pc, x, y, w, h]
  They represent classes(20), Probability x,y,w,h(Box)
* Prediction_cell = [c1, c2, c3, ..... c20, Pc, x, y, w, h, Pc, x, y, w, h]
  They represent classes(20), Probability x,y,w,h(Box1), Probability x,y,w,h(Box2)
