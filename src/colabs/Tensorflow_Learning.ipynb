Basic Neural Network

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data() 
```

```python
print(x_train.shape)
print(y_train.shape)

x_train = tf.reshape(x_train, (-1, 784))
print(x_train.shape)

x_test = tf.reshape(x_test, (-1, 784))
print(x_test.shape)
```

```python
x_train = tf.cast(x_train, dtype=tf.float32)
x_test = tf.cast(x_test, dtype=tf.float32)
```

```python
x_train = x_train / 255.0 
x_test = x_test / 255.0
```

```python
#Sequential API - Not flexible
model = keras.Sequential(
    [
      keras.Input(28*28),
      layers.Dense(512, activation='relu'),
      layers.Dense(256, activation='relu'),
      layers.Dense(10), # Softmax NOT present so from_logits = True
           
    ]
)

print(model.summary()) # Needed with an input or model.compile and model.fit


inputs = keras.Input(28*28)
x = layers.Dense(512, activation='relu')(inputs)
x = layers.Dense(256, activation='relu')(x)
outputs= layers.Dense(10, activation='softmax')(x) # Softmax present so from_logits = False
model = keras.Model(inputs=inputs, outputs=outputs)

model.compile(
    loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True),  # Sparse -> No one hot encoding 
    optimizer = keras.optimizers.Adam(learning_rate=0.001),
    metrics = ["accuracy"],
)

model.fit(
    x_train, y_train, batch_size=32, epochs=5, verbose=2
)

model.evaluate(
    x_test, y_test, batch_size=32, verbose=2
)
```
